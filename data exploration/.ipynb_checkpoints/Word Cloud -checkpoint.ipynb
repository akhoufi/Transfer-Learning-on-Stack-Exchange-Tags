```python
import csv
import operator
from collections import defaultdict
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
dataframes = {
    "cooking": pd.read_csv("../../input_light/cooking_light.csv"),
    "crypto": pd.read_csv("../../input_light/crypto_light.csv"),
    "robotics": pd.read_csv("../../input_light/robotics_light.csv"),
    "biology": pd.read_csv("../../input_light/biology_light.csv"),
    "travel": pd.read_csv("../../input_light/travel_light.csv"),
    "diy": pd.read_csv("../../input_light/diy_light.csv"),
    "physics": pd.read_csv("../../input_light/physics_light.csv"),
}
```

## Word Cloud representation of Title and Content

```python
for name, df in dataframes.items():
    text = df['title'].str.cat(sep=' ') 
    wc_title = WordCloud(max_words=1000,
               random_state=1).generate(text)
    plt.figure()
    plt.imshow(wc_title)
    plt.axis("off")
    plt.title(name +' Titles Word Cloud')
    plt.show()
```

```python
for name, df in dataframes.items():
    text = df['content'].str.cat(sep=' ') 
    wc_content = WordCloud(max_words=1000,
               random_state=1).generate(text)
    plt.figure()
    plt.imshow(wc_content)
    plt.axis("off")
    plt.title(name +' Content Word Cloud')
    plt.show()
```

```python
# Easyer processing of tags
for name, df in dataframes.items():
    if(name != "physics"):
        print(name)
        df["tags"] = df["tags"].replace(r"\[(.*)\]", r"\1",regex=True).str.replace("'","")
```

```python
tagsDict = {}
for name, df in dataframes.items():
    if(name != "physics"):
        print(name)
        #Dictionnaire des tag
        tagsDict[name] = {}
        for tags in df['tags']:
            tagList = tags.split(',')
            for tag in tagList:
                if (tag in tagsDict[name]):
                    tagsDict[name][tag] += 1
                else:
                    tagsDict[name][tag] = 1
                    
        print(len(tagsDict[name]))
        biggestTag = max(tagsDict[name], key=tagsDict[name].get)
        print("Most popular tag: " + str(biggestTag) + " with "+ str(tagsDict[name][biggestTag]) + " occurences.")
    '''print(df["tags"][0])
    tags = df["tags"][0][1:-1]
    tags = tags.replace("'","")
    tags = tags.split(",")'''
    #print(tags)
```
